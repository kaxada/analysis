# Loading needed libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
import numpy as np
from scipy.stats import shapiro
from scipy.stats import levene
from scipy.stats import kruskal
import scikit_posthocs as sp

plt.style.use('seaborn-darkgrid')
sns.set_context('notebook')
```

# Reading tables

```python
# Reading CSV files
label_table = pd.read_csv("../../data/label_dataset.csv", usecols=lambda column: column != 'Unnamed: 0')
print(label_table.dtypes)  

commiters_18_time = pd.read_csv("committer_time.csv", usecols=lambda column: column != 'Unnamed: 0')
print(commiters_18_time.dtypes)

# Joining dataframes
commiters_time_label = pd.merge(commiters_18_time, label_table, on="repo_id", how="inner")
print(commiters_time_label.dtypes)

# Selecting specific columns
commiterTimeSelectedVar = commiters_time_label[["repo_id", "committer_count", "label_name"]]
```

# General Visualization
## Excluding outliers

```python
# Exclude outliers
def exclude_outliers(data, variable, group_var, sd_threshold=2):
    # Function to filter outliers in a group
    def filter_group(group):
        mean_val = group[variable].mean()
        sd_val = group[variable].std()
        return group[np.abs(group[variable] - mean_val) < sd_threshold * sd_val]
    
    # Apply the filter function to each group and combine the results
    return data.groupby(group_var).apply(filter_group).reset_index(drop=True)
```

```python
commiterTimeSelectedVar_filtered = exclude_outliers(commiterTimeSelectedVar, "committer_count", "label_name", sd_threshold=2)
```

## Calculating descriptive stats showing on the plot

```python
# Calculate stats to show on the plot
stats_df = commiterTimeSelectedVar_filtered.groupby('label_name')['committer_count'].agg(
    average='mean',
    upper_quartile=lambda x: x.quantile(0.75)
).reset_index()

stats_df['label_pos'] = stats_df['upper_quartile'] + 0.5
```

## Generating box plot

```python
# Set plots style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# Create a boxplot
ax = sns.boxplot(x='label_name', y='committer_count', data=commiterTimeSelectedVar_filtered, palette="Set2", dodge=True)

# Add mean values as text on the plot
for index, row in stats_df.iterrows():
    ax.text(
        x=index, 
        y=row['label_pos'], 
        s=f"Mean: {row['average']:.2f}", 
        color='black', 
        ha="center", 
        size=12
    )

# Set plot title and labels
ax.set_title('Box Plot with Descriptive Statistics', fontsize=14)
ax.set_xlabel('Label Name', fontsize=12)
ax.set_ylabel('Commiters', fontsize=12)

# Making plot easier to read
sns.despine(trim=True)
plt.xticks(rotation=45)

plt.show()
```

## Kruskal-Wallis H test to test whether there is a significant difference among pairs of groups

```python
# Preparing data for Kruskal-Wallis H Test
groups = [group['committer_count'].values for name, group in commiterTimeSelectedVar.groupby('label_name')]

# Perform Kruskal-Wallis H Test
stat, p = kruskal(*groups)
print(f'Kruskal-Wallis H Test Statistic: {stat}, P-value: {p}')
```

## Dunn's test to conduct post-hoc test for examining the differience between each pair of groups.

```python
# Conduct Dunn's Test
dunn_test_result = sp.posthoc_dunn(commiterTimeSelectedVar, val_col='committer_count', group_col='label_name', p_adjust='fdr_bh')

print(dunn_test_result)
```
